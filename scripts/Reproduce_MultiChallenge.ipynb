```python
import json, subprocess
from pathlib import Path
from collections import defaultdict

REPO = "/content/multi-challenge"
RESP_DIR = f"{REPO}/data/final_model_responses"
OUT_DIR  = f"{REPO}/outputs"
os.makedirs(OUT_DIR, exist_ok=True)

files = [
    "gpt-4o-2024-08-06_responses.jsonl",
    "claude-3-5-sonnet-20241022_responses.jsonl",
    "gemini-1.5-pro-002_responses.jsonl",
    "llama3-1-405b-instruct-v1_responses.jsonl",
    "mistral-large-latest_responses.jsonl",
    "o1-preview_responses.jsonl",
]

ATTEMPTS = 1
WORKERS  = 1

# IMPORTANT: to avoid re-paying, default to reuse existing outputs
REUSE_IF_EXISTS = True

# Only relevant if ATTEMPTS > 1
# "first" aligns best with paper-style single-attempt reporting
AGGREGATION = "first"  # one of: "first", "or", "and", "avg"

axes_order = ["INSTRUCTION_RETENTION", "INFERENCE_MEMORY", "RELIABLE_VERSION_EDITING", "SELF_COHERENCE"]

def compute_scores_from_results_list(results_list, aggregation="first"):
    """
    results_list: list of dicts with fields: axis, question_id, passed(bool), attempt(int)
    aggregation:
      - first: use attempt==0 only
      - or: question passes if any attempt passes
      - and: question passes if all attempts pass
      - avg: per-question pass rate averaged (still reported as %)
    """
    by_axis_qid = defaultdict(list)  # (axis, qid) -> list[(attempt, passed)]
    for r in results_list:
        axis = r.get("axis")
        qid  = r.get("question_id")
        if axis is None or qid is None:
            continue
        passed = bool(r.get("passed", False))
        attempt = r.get("attempt", 0)
        # attempt may be "NA" on errors; skip those
        if isinstance(attempt, str):
            continue
        by_axis_qid[(axis, qid)].append((attempt, passed))

    axis_tot = defaultdict(int)
    axis_pass = defaultdict(float)

    for (axis, qid), ap_list in by_axis_qid.items():
        axis_tot[axis] += 1
        ap_list_sorted = sorted(ap_list, key=lambda x: x[0])
        passed_list = [p for _, p in ap_list_sorted]

        if aggregation == "first":
            # use attempt 0 if present, else fallback to first available
            p = passed_list[0] if passed_list else False
            axis_pass[axis] += 1.0 if p else 0.0
        elif aggregation == "or":
            axis_pass[axis] += 1.0 if any(passed_list) else 0.0
        elif aggregation == "and":
            axis_pass[axis] += 1.0 if (len(passed_list) > 0 and all(passed_list)) else 0.0
        elif aggregation == "avg":
            axis_pass[axis] += (sum(1.0 for p in passed_list if p) / max(1, len(passed_list)))
        else:
            raise ValueError(f"Unknown aggregation: {aggregation}")

    axis_scores = {ax: (axis_pass[ax] / axis_tot[ax] * 100.0) for ax in axis_tot}
    overall = sum(axis_scores.values()) / len(axis_scores) if axis_scores else None
    return overall, axis_scores

def load_scores_from_outjson(out_json_path, attempts, aggregation):
    """
    Try to read scores from out_json. If not present, compute from raw results.
    """
    with open(out_json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    # 1) If already includes "scores"
    if isinstance(data, dict):
        scores = data.get("scores") or data.get("SCORES") or data.get("summary")
        if isinstance(scores, dict) and ("overall_score" in scores) and ("axis_scores" in scores):
            return float(scores["overall_score"]), dict(scores["axis_scores"])

        # 2) dict wrapping a list of results
        for key in ["evaluation_results", "results", "raw_results"]:
            if key in data and isinstance(data[key], list):
                return compute_scores_from_results_list(data[key], aggregation=("first" if attempts == 1 else aggregation))

    # 3) top-level list of results
    if isinstance(data, list):
        return compute_scores_from_results_list(data, aggregation=("first" if attempts == 1 else aggregation))

    return None, None

def run_one(resp_file: str):
    base = resp_file.replace(".jsonl", "")
    resp_path = f"{RESP_DIR}/{resp_file}"

    # include attempts in filename to avoid reading stale results
    out_json = f"{OUT_DIR}/{base}_judge_results_a{ATTEMPTS}.json"
    out_csv  = f"{OUT_DIR}/{base}_judge_results_a{ATTEMPTS}.csv"

    if not Path(resp_path).exists():
        return {"model": base, "status": "MISSING_FILE"}

    # If reuse enabled and output exists, do NOT rerun (saves money)
    if REUSE_IF_EXISTS and Path(out_json).exists():
        overall, axis_scores = load_scores_from_outjson(out_json, ATTEMPTS, AGGREGATION)
        if overall is None or axis_scores is None:
            raise RuntimeError(f"Found {out_json} but could not parse scores. Delete it or set REUSE_IF_EXISTS=False.")
        return {"model": base, "status": "OK(CACHED)", "overall": overall, "axis_scores": axis_scores, "out_json": out_json, "out_csv": out_csv}

    # Safety: if attempts>1 and aggregation isn't explicit, stop.
    if ATTEMPTS > 1 and AGGREGATION not in ["first", "or", "and", "avg"]:
        raise ValueError("ATTEMPTS>1 requires AGGREGATION in {first, or, and, avg}")

    cmd = [
        "python", "-m", "run_judge_eval",
        "--responses", resp_path,
        "--out_json", out_json,
        "--out_csv",  out_csv,
        "--workers",  str(WORKERS),
        "--attempts", str(ATTEMPTS),
    ]

    print(f"\n=== Running: {base} (attempts={ATTEMPTS}, workers={WORKERS}) ===")
    p = subprocess.run(cmd, cwd=REPO, capture_output=True, text=True)
    print(p.stdout)
    if p.returncode != 0:
        print("---- STDERR ----")
        print(p.stderr)
        raise RuntimeError(f"run_judge_eval failed for {base} with code {p.returncode}")

    if not Path(out_json).exists():
        raise RuntimeError(f"Expected output json not found: {out_json}")

    overall, axis_scores = load_scores_from_outjson(out_json, ATTEMPTS, AGGREGATION)
    if overall is None or axis_scores is None:
        raise RuntimeError(f"Could not parse scores from: {out_json}")

    return {"model": base, "status": "OK", "overall": overall, "axis_scores": axis_scores, "out_json": out_json, "out_csv": out_csv}

# Run all
results = [run_one(f) for f in files]

# Print summary table
print("\n================ SUMMARY ================")
print("\t".join(["model", "status", "overall"] + axes_order))
for r in results:
    row = [r["model"], r["status"]]
    ov = r.get("overall")
    row.append("" if ov is None else f"{ov:.4f}")
    d = r.get("axis_scores") or {}
    for ax in axes_order:
        v = d.get(ax)
        row.append("" if v is None else f"{v:.4f}")
    print("\t".join(row))

# Save summary (Drive if mounted, else outputs/)
drive_dir = "/content/drive/MyDrive/multichallenge"
summary_path = f"{drive_dir}/all_models_judge_summary.json" if Path("/content/drive/MyDrive").exists() else f"{OUT_DIR}/all_models_judge_summary.json"
if Path("/content/drive/MyDrive").exists():
    os.makedirs(drive_dir, exist_ok=True)

with open(summary_path, "w", encoding="utf-8") as f:
    json.dump(results, f, indent=2)
print(f"\nSaved summary: {summary_path}")

```
